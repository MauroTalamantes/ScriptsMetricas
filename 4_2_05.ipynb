# 4_2_05 - Representaci√≥n de organizaciones privadas con posible conflicto de inter√©s
| **Escenario**                                                          | **Condici√≥n exacta evaluada**                                                                      | **Resultado**    |
| ---------------------------------------------------------------------- | -------------------------------------------------------------------------------------------------- | ---------------- |
| **No hay representaciones declaradas**                                 | La lista `representaciones.representacion` no existe o est√° vac√≠a                                  | ‚ö™ **NO_APLICA**  |
| **Faltan datos esenciales**                                            | No existe:<br>‚Ä¢ nombre del ente p√∫blico, o <br>‚Ä¢ sector productivo de la organizaci√≥n representada | ‚ö™ **SIN_DATO**   |
| **La organizaci√≥n representada pertenece a sector de riesgo T, D o C** | `sector ‚àà sectores_riesgo[funcion_principal_del_ente]`                                             | üî¥ **NO_CUMPLE** |
| **La organizaci√≥n NO pertenece a sectores de riesgo**                  | `sector` **no** est√° dentro de los sectores riesgo asociados al ente p√∫blico                       | üü¢ **CUMPLE**    |
| **Registros existentes pero incompletos**                              | Sectores vac√≠os, sin clasificar o nulos                                                            | ‚ö™ **SIN_DATO**   |


```python
import traceback
from pymongo import MongoClient, UpdateOne
from config import MONGO_URI, DB_NAME, SOURCE_COLLECTION_NAME, METRICS_COLLECTION_NAME
import unicodedata


METRIC_ID = "4_2_05_REPRESENTACION_SECTOR_RIESGO"


# -------------------------------------------------------------------
# MATRIZ TEMPORAL (sustituir por cat√°logo INEGI real)
# -------------------------------------------------------------------
MATRIZ_SECTORES_FUNCIONES = {
    "SALUD": ["T", "D"],
    "OBRAS PUBLICAS": ["T", "C"],
    "EDUCACION": ["D"],
    "SEGURIDAD": ["T", "C"],
}


# -------------------------------------------------------------------
# AUXILIARES
# -------------------------------------------------------------------
def limpiar(v):
    if not v:
        return None
    v = str(v).strip().upper()
    v = ''.join(c for c in unicodedata.normalize('NFD', v)
                if unicodedata.category(c) != 'Mn')  # sin acentos
    while "  " in v:
        v = v.replace("  ", " ")
    return v


def extraer(doc, path):
    try:
        partes = path.split(".")
        actual = doc
        for p in partes:
            if isinstance(actual, dict):
                actual = actual.get(p)
            elif isinstance(actual, list) and len(actual) > 0:
                actual = actual[0].get(p)
            else:
                return None
        return actual
    except:
        return None


# -------------------------------------------------------------------
# EVALUADOR DE LA M√âTRICA
# -------------------------------------------------------------------
def evaluar_metrica(doc):

    # A) Ente p√∫blico donde labora
    ente = limpiar(
        extraer(doc, "declaracion.situacionPatrimonial.datosEmpleoCargoComision.nombreEntePublico")
    )

    if not ente:
        return "SIN_DATO"

    funcion_principal = ente  # se reemplaza al integrar cat√°logo real
    sectores_riesgo = MATRIZ_SECTORES_FUNCIONES.get(funcion_principal, [])

    # B) Representaciones declaradas
    reps = extraer(doc, "declaracion.intereses.representaciones.representacion")

    if not reps:
        return "NO_APLICA"

    if not isinstance(reps, list):
        reps = [reps]

    hubo_datos = False

    for rep in reps:
        sector = limpiar(rep.get("sector"))
        tipo = limpiar(rep.get("tipo"))  # declarante/pareja/dependiente si existe

        if not sector:
            return "SIN_DATO"

        hubo_datos = True

        # Regla principal
        if sector in sectores_riesgo:
            return "NO_CUMPLE"

    if not hubo_datos:
        return "NO_APLICA"

    return "CUMPLE"


# -------------------------------------------------------------------
# PROCESADOR MONGO
# -------------------------------------------------------------------
def procesar_metrica_4_2_05():
    resultados = {"CUMPLE": 0, "NO_CUMPLE": 0, "SIN_DATO": 0, "NO_APLICA": 0}
    operaciones = []
    total = 0

    try:
        print(f"\nüöÄ Procesando m√©trica {METRIC_ID}...\n")

        client = MongoClient(MONGO_URI, serverSelectionTimeoutMS=5000)
        db = client[DB_NAME]

        src = db[SOURCE_COLLECTION_NAME]
        tgt = db[METRICS_COLLECTION_NAME]

        cursor = src.find({}, {
            "_id": 1,
            "declaracion.situacionPatrimonial.datosEmpleoCargoComision": 1,
            "declaracion.intereses.representaciones": 1
        }, no_cursor_timeout=True)

        for doc in cursor:
            total += 1

            try:
                resultado = evaluar_metrica(doc)
            except Exception as e:
                resultado = "SIN_DATO"
                print(f"‚ö†Ô∏è Error en documento {doc.get('_id')}: {e}")

            resultados[resultado] += 1

            operaciones.append(
                UpdateOne(
                    {"_id": doc["_id"]},
                    {"$set": {METRIC_ID: resultado}},
                    upsert=True
                )
            )

            # Bulk cada 2000
            if len(operaciones) >= 2000:
                tgt.bulk_write(operaciones)
                operaciones.clear()
                print(f" > {total} documentos procesados...")

        # √öltimo bulk
        if operaciones:
            tgt.bulk_write(operaciones)

        print("\n--- RESUMEN FINAL ---")
        print("Total procesados:", total)
        for k, v in resultados.items():
            print(f"  {k}: {v}")

    except:
        traceback.print_exc()

    finally:
        try:
            client.close()
        except:
            pass

        print("Conexi√≥n cerrada.")


if __name__ == "__main__":
    procesar_metrica_4_2_05()
# 
```
