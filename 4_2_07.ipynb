# Clientes principales (personas morales) con posible conflicto de inter√©s ‚Äî M√©trica 4_2_07
| **Escenario**                                                     | **Condici√≥n evaluada**                                         | **Resultado**    |
| ----------------------------------------------------------------- | -------------------------------------------------------------- | ---------------- |
| **No hay clientes principales declarados**                        | No existe la lista `clientesPrincipales.cliente` o viene vac√≠a | ‚ö™ **NO_APLICA**  |
| **Cliente no es persona moral / no aplica la evaluaci√≥n**         | Campo ‚Äúotorgante‚Äù no indica persona moral                      | ‚ö™ **NO_APLICA**  |
| **Faltan datos clave**                                            | Sector productivo vac√≠o o inexistente                          | ‚ö™ **SIN_DATO**   |
| **Sector productivo pertenece a categor√≠as de riesgo (T, D o C)** | `sector ‚àà sectores_riesgo[funcion_principal_del_ente]`         | üî¥ **NO_CUMPLE** |
| **Sector NO pertenece a sectores riesgosos**                      | Sector no est√° en T, D o C                                     | üü¢ **CUMPLE**    |


```python
import traceback
from pymongo import MongoClient, UpdateOne
from config import MONGO_URI, DB_NAME, SOURCE_COLLECTION_NAME, METRICS_COLLECTION_NAME
import unicodedata


METRIC_ID = "4_2_07_CLIENTES_PRINCIPALES_SECTOR_RIESGO"


# -------------------------------------------------------------------
# MATRIZ TEMPORAL (se sustituir√° por cat√°logo INEGI real)
# -------------------------------------------------------------------
MATRIZ_SECTORES_FUNCIONES = {
    "SALUD": ["T", "D"],
    "OBRAS PUBLICAS": ["T", "C"],
    "EDUCACION": ["D"],
    "SEGURIDAD": ["T", "C"],
}


# -------------------------------------------------------------------
# AUXILIARES
# -------------------------------------------------------------------
def limpiar(v):
    if not v:
        return None
    v = str(v).strip().upper()
    v = ''.join(c for c in unicodedata.normalize('NFD', v)
                if unicodedata.category(c) != 'Mn')
    while "  " in v:
        v = v.replace("  ", " ")
    return v


def extraer(doc, path):
    try:
        partes = path.split(".")
        actual = doc
        for p in partes:
            if isinstance(actual, dict):
                actual = actual.get(p)
            elif isinstance(actual, list) and len(actual) > 0:
                actual = actual[0].get(p)
            else:
                return None
        return actual
    except:
        return None


# -------------------------------------------------------------------
# EVALUADOR DE M√âTRICA
# -------------------------------------------------------------------
def evaluar_metrica(doc):

    # A) Ente p√∫blico
    ente = limpiar(
        extraer(doc, "declaracion.situacionPatrimonial.datosEmpleoCargoComision.nombreEntePublico")
    )

    if not ente:
        return "SIN_DATO"

    funcion_principal = ente
    sectores_riesgo = MATRIZ_SECTORES_FUNCIONES.get(funcion_principal, [])

    # B) Clientes principales
    clientes = extraer(doc, "declaracion.intereses.clientesPrincipales.cliente")

    if not clientes:
        return "NO_APLICA"

    if not isinstance(clientes, list):
        clientes = [clientes]

    hubo_datos = False

    for c in clientes:

        otorgante = limpiar(c.get("otorgante"))
        sector = limpiar(c.get("sector"))

        # Solo se eval√∫an personas morales
        if otorgante != "PERSONA MORAL":
            continue

        if not sector:
            return "SIN_DATO"

        hubo_datos = True

        # Regla principal de conflicto
        if sector in sectores_riesgo:
            return "NO_CUMPLE"

    if not hubo_datos:
        return "NO_APLICA"

    return "CUMPLE"


# -------------------------------------------------------------------
# PROCESADOR MONGO
# -------------------------------------------------------------------
def procesar_metrica_4_2_07():
    resultados = {"CUMPLE": 0, "NO_CUMPLE": 0, "SIN_DATO": 0, "NO_APLICA": 0}
    operaciones = []
    total = 0

    try:
        print(f"\nüöÄ Procesando m√©trica {METRIC_ID}...\n")

        client = MongoClient(MONGO_URI, serverSelectionTimeoutMS=5000)
        db = client[DB_NAME]

        src = db[SOURCE_COLLECTION_NAME]
        tgt = db[METRICS_COLLECTION_NAME]

        cursor = src.find({}, {
            "_id": 1,
            "declaracion.situacionPatrimonial.datosEmpleoCargoComision": 1,
            "declaracion.intereses.clientesPrincipales": 1
        }, no_cursor_timeout=True)

        for doc in cursor:
            total += 1

            try:
                resultado = evaluar_metrica(doc)
            except Exception as e:
                resultado = "SIN_DATO"
                print(f"‚ö†Ô∏è Error documento {doc.get('_id')}: {e}")

            resultados[resultado] += 1

            operaciones.append(
                UpdateOne(
                    {"_id": doc["_id"]},
                    {"$set": {METRIC_ID: resultado}},
                    upsert=True
                )
            )

            if len(operaciones) >= 2000:
                tgt.bulk_write(operaciones)
                operaciones.clear()
                print(f" > {total} documentos procesados...")

        if operaciones:
            tgt.bulk_write(operaciones)

        print("\n--- RESUMEN FINAL ---")
        print("Total procesados:", total)
        for k, v in resultados.items():
            print(f"  {k}: {v}")

    except:
        traceback.print_exc()

    finally:
        try:
            client.close()
        except:
            pass

        print("Conexi√≥n cerrada.")


if __name__ == "__main__":
    procesar_metrica_4_2_07()
```
